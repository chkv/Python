```python
class Node:
    def __init__(self,data,next=None):
        self.data=data
        self.next=next
```

```python
class LinkedList:
    def __init__ (self):
        self.head=None

    def insert_at_beginning(self,data):
        node=Node(data,self.head)
        self.head=node

    def print(self):
        if self.head is None:
            print("Empty Linked List")
        itr=self.head
        lstr=""
        while itr:
            lstr+=str(itr.data)+" ==> "
            itr=itr.next
        print(lstr)

    def insert_end(self,data):
        if(self.head is None):
            self.head=Node(data)
            return

        itr=self.head
        while itr.next:
            itr=itr.next
        itr.next=Node(data)

    def get_length(self):
        count=0
        itr=self.head
        while itr:
            count+=1
            itr=itr.next
        return count

    def insert_at(self,index,data):

        if index<0 or index>self.get_length():
            raise Exception("index out of range")

        if index==0:
            self.insert_at_beginning(data)
            return
        count=0
        itr=self.head
        while itr:
            if count==index-1:
                itr.next=Node(data,itr.next)
                break
            itr=itr.next
            count+=1

    def delete_at(self,index):
        if index<0 or index>=self.get_length():
            raise Exception("index out of range")

        if index==0:
            self.head=self.head.next
            return

        count=0
        itr=self.head

        while itr:
            if count==index-1:
                itr.next=itr.next.next
                return

            itr=itr.next
            count+=1

    def insert(self,list):
        self.head=None
        for l in list:
            self.insert_end(l)

    def insert_after_value(self,data_next,data_new):
        itr=self.head
        index=0
        while itr:
            if itr.data==data_next:
                self.insert_at(index+1,data_new)
                return
            index+=1
            itr=itr.next

        raise Exception("Value not found")

    def remove_by_value(self,value):
        if self.head is None:
            raise Exception("Empty Linked List")

        if self.head.data==value:
            self.head=self.head.next
            return

        itr=self.head
        while itr.next:
            if itr.next.data==value:
                itr.next=itr.next.next
                return
            itr=itr.next
        raise Exception("Value not found")



```

```python

ll = LinkedList()
ll.insert(["banana","mango","grapes","orange"])
ll.print()
```

```python
ll.insert_after_value("mango","apple") # insert apple after mango
ll.print()
```

```python
ll.remove_by_value("orange") # remove orange from linked list
ll.print()
```

```python
ll.remove_by_value("figs")
ll.print()
```

```python
ll.remove_by_value("banana")
ll.remove_by_value("mango")
ll.remove_by_value("apple")
ll.remove_by_value("grapes")
ll.print()
```
