```python
class NodeTree:
    def __init__(self,value):
        self.value = value
        self.parent=None
        self.child=[]
    def add_child(self,child):
        child.parent=self
        self.child.append(child)
    def level(self):
        level=0
        p=self.parent
        while p:
            p=p.parent
            level=level+1
        return level
    def print(self,ind=3):
        if self.level()>ind:
            return
        prefix=" "*self.level()*3+"|__" if self.parent else ""
        print(f"{prefix}{self.value}")
        for child in self.child:
            child.print(ind)

```

```python
def locations():
    node=NodeTree("Global")
    india=NodeTree("India")
    usa=NodeTree("USA")
    node.add_child(india)
    node.add_child(usa)
    gujarat=NodeTree("Gujarat")
    karnataka=NodeTree("Karnataka")
    india.add_child(gujarat)
    india.add_child(karnataka)
    ahmedabad=NodeTree("Ahmedabad")
    baroda=NodeTree("Baroda")
    gujarat.add_child(ahmedabad)
    gujarat.add_child(baroda)
    bangalore=NodeTree("Bangalore")
    mysore=NodeTree("Mysore")
    karnataka.add_child(bangalore)
    karnataka.add_child(mysore)
    newjersey=NodeTree("Newjersey")
    california=NodeTree("California")
    usa.add_child(newjersey)
    usa.add_child(california)
    princeton=NodeTree("Princeton")
    trenton=NodeTree("Trenton")
    newjersey.add_child(princeton)
    newjersey.add_child(trenton)
    sanfrancisco=NodeTree("Sanfrancisco")
    mountasinview=NodeTree("Mountasinview")
    california.add_child(mountasinview)
    california.add_child(sanfrancisco)
    return node
```

```python
couintries=locations()
couintries.print(3)
```
