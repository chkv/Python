```python
from itertools import count


class Node:
    def __init__(self,data,next=None,previous=None):
        self.data=data
        self.next=next
        self.previous=previous
```

```python
class DoubleLinkedList:
    def __init__(self):
        self.head=None

    def insert_at_beginning(self,data):
        if self.head is None:
            self.head=Node(data)
            return
        itr=self.head
        node=Node(data,itr)
        itr.previous=node
        self.head=node

    def insert_at_end(self,data):
        if self.head is None:
            self.head=Node(data)
            return
        itr=self.head
        while itr.next:
            itr=itr.next
        itr.next=Node(data,None,itr)

    def print_forward(self):
        if self.head is None:
            raise Exception("Empty Linked List")
        itr=self.head

        dl_str=""
        itr=self.head
        while itr:
            dl_str+=str(itr.data)+ " ==> "
            itr=itr.next
        print(dl_str)

    def print_backward(self):
        if self.head is None:
            raise Exception("Empty Linked List")
        itr_last=self.head
        while itr_last.next :
            itr_last=itr_last.next
        dl_str=""
        while itr_last :
            dl_str+=str(itr_last.data)+ " ==> "
            itr_last=itr_last.previous
        print(f"Reverse Order: {dl_str}")

    def get_length(self):
        itr=self.head
        count=0
        while itr:
            count+=1
            itr=itr.next
        return count
        print(f"Length: {count}")

    def insert_at(self,index,value):
        if index<0 or self.get_length()<index:
            raise Exception("Index out of range")
        if index==0:
            self.insert_at_beginning(value)
            return
        count=0
        itr=self.head
        while itr:
            if count==index-1:
                node=Node(value,itr.next,itr)
                if node.next:
                    node.next.previous=node
                itr.next=node
                break
            count+=1
            itr=itr.next

    def delete_at(self,index):
        if index<0 or self.get_length()<=index:
            raise Exception("Index out of range")
        if index==0:
            self.head=self.head.next
            if self.head:
                self.head.previous=None
            return
        count=0
        itr=self.head
        while itr:
            if count==index-1:
                itr.next=itr.next.next
                if itr.next :
                    itr.next.previous=itr
                return
            count+=1
            itr=itr.next

    def insert(self,list):
        self.head=None
        for data in list:
            self.insert_at_end(data)


```

```python
dl=DoubleLinkedList()
dl.insert_at_beginning(2)
dl.insert_at_end(3)
dl.insert_at_beginning(1)
dl.print_forward()
dl.print_backward()
# Expected: Reverse Order: 3 ==> 2 ==> 1 ==>
```

```python
dl.insert_at(1, 5)
dl.print_forward()

# Expected:
# 1 ==> 5 ==> 2 ==> 3 ==>
```

```python
dl.insert_at(dl.get_length(), 9)
dl.print_forward()

# Expected:
# 1 ==> 5 ==> 2 ==> 3 ==> 9 ==>
```

```python
dl.delete_at(2)
dl.print_forward()

# Expected:
# 1 ==> 5 ==> 3 ==> 9 ==>
```

```python
dl.delete_at(0)
dl.print_forward()

# Expected:
# 5 ==> 3 ==> 9 ==>
```

```python
dl.delete_at(dl.get_length()-1)
dl.print_forward()

# Expected:
# 5 ==> 3 ==>
```

```python
dl.delete_at(0)
dl.delete_at(0)
dl.print_forward()

# Should raise: Empty Linked List
```
