```python
class Tree:
    def __init__(self,routes):
        self.routes=routes
        self.graph_dict={}

        for start,end in routes:
            if start in self.graph_dict:
                self.graph_dict[start].append(end)
            else:
                self.graph_dict[start]=[end]

        print(f"graph dict is",self.graph_dict)

    def get_routes(self,start,end,path=[]):
        path=path+[start]
        if start==end:
            return [path]
        if start not in self.graph_dict:
            return []
        paths=[]
        for node in self.graph_dict[start]:
            if node not in path:
                new_path=self.get_routes(node,end,path)
                for p in new_path:
                    paths.append(p)
        return paths


```

```python
routes = [
        ("Mumbai","Pune"),
        ("Mumbai", "Surat"),
        ("Surat", "Bangaluru"),
        ("Pune","Hyderabad"),
        ("Pune","Mysuru"),
        ("Hyderabad","Bangaluru"),
        ("Hyderabad", "Chennai"),
        ("Mysuru", "Bangaluru"),
        ("Chennai", "Bangaluru")
]
```

```python
g=Tree(routes)
```
