Binary Tree Part 1 Exercise
Add following methods to BinarySearchTreeNode class created in main video tutorial

1. find_min(): finds minimum element in entire binary tree
2. find_max(): finds maximum element in entire binary tree
3. calculate_sum(): calcualtes sum of all elements
4. post_order_traversal(): performs post order traversal of a binary tree
5. pre_order_traversal(): perofrms pre order traversal of a binary tree

```python
class Nodetree:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
    def add_child(self, value):
        if value == self.data:
            raise Exception(f"{value}: Duplicates not allowed")

        if value < self.data:
            if self.left:
                self.left.add_child(value)
            else:
                self.left = Nodetree(value)

        else:  # value > self.data
            if self.right:
                self.right.add_child(value)
            else:
                self.right = Nodetree(value)

    def print_inorder(self):
        element=[]
        if self.left:
            element+=self.left.print_inorder()
        element.append(self.data)
        if self.right:
            element+=self.right.print_inorder()
        return element

    def print_postorder(self):
        element=[]
        if self.left:
            element+=self.left.print_postorder()
        if self.right:
            element+=self.right.print_postorder()
        element.append(self.data)
        return element
    def print_preorder(self):
        element=[]
        element.append(self.data)
        if self.left:
            element+=self.left.print_preorder()
        if self.right:
            element+=self.right.print_preorder()
        return element
    def min_element(self):
        if self.left:
            return self.left.min_element()
        return self.data

    def max_element(self):
        if self.right:
            return self.right.max_element()
        return self.data

    def calc(self):

        list_sum=self.print_inorder()
        return sum(list_sum)

```

```python
def build_binary_tree():
    list=[16,21,3,12,34,20,4,18,15,31]

    node=Nodetree(list[0])
    for i in range(1,len(list)):
        node.add_child(list[i])
    return node
```

```python
n=build_binary_tree()
print(n.print_inorder())
print(n.print_preorder())
print(n.print_postorder())
```

```python
print(n.min_element())
```

```python
print(n.max_element())
```

```python
print(n.calc())
```
