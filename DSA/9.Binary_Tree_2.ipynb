```python
class BinaryTree:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
    def add_child(self,value):
        if self.data==value:
            return
        if value<self.data:
            if self.left:
                self.left.add_child(value)
            else:
                self.left=BinaryTree(value)
        if value>self.data:
            if self.right:
                self.right.add_child(value)
            else:
                self.right=BinaryTree(value)

    def print_inorder(self):
        element=[]
        if self.left:
            element+=self.left.print_inorder()
        element.append(self.data)
        if self.right:
            element+=self.right.print_inorder()
        return element

    def min_value(self):
        if self.left:
            return self.left.min_value()
        return self.data
    def delete(self, value):

        if value < self.data:
            if self.left:
                self.left = self.left.delete(value)

        elif value > self.data:
            if self.right:
                self.right = self.right.delete(value)

        else:
            # Case 1: No children
            if self.left is None and self.right is None:
                return None

            # Case 2: One child
            if self.left is None:
                return self.right

            if self.right is None:
                return self.left

            # Case 3: Two children
            max_value = self.left.max_value()
            self.data = max_value
            self.left = self.left.delete(max_value)

        return self
```

```python
def build_binary(list):
    tree=BinaryTree(list[0])
    for i in range(1,len(list)):
        tree.add_child(list[i])
    return tree
```

```python
list=[16,21,3,12,34,20,4,18,15,31]
bt=build_binary(list)
bt.print_inorder()
```

```python
bt.delete(20)
bt.print_inorder()
```
