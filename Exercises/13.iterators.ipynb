**#Exercise: Iterators**
- Create an iterator for fibonacci series in such a way that each next returns the next element from fibonacci series.
 - The iterator should stop when it reaches a limit defined in the constructor.

```python



class Fibonacci:
    def __init__(self,limit):
        self.limit = limit
        self.previous=0
        self.current=1
        self.counter=1
    def __iter__(self):
        return self
    def __next__(self):
        if self.counter<self.limit:
            result=self.previous+self.current
            self.counter+=1
            self.previous=self.current
            self.current=result
            return result
        else:
            raise StopIteration

fib=Fibonacci(10)
itr=iter(fib)
while itr:
    try:
        print(next(itr))
    except StopIteration:
        break


```
