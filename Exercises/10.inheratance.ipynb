```python
class Animal:
    def usual_comparison(self):
        print("Less Intelligence compared to human")
    def sound(self):
        print("Rawr!")

class Dog(Animal):
    def sound(self):
        super().sound()
        print("Woof!")

d=Dog()
d.usual_comparison()
d.sound()
```
