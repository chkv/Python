Print Square Sequence using yield
    Create Generator method such that every time it will returns a next square number

for exmaple : 1 4 9 16 ..


```python
def sqr():
    x=1
    while True:
       yield x**2
       x+=1
# for i in sqr():
#     if i>40:
#         break
#     print(i)

p=sqr()
print(next(p))
print(next(p))
print(next(p))
print(next(p))
```

```python
def gen():
    yield 1
    yield 2
    yield 3
    yield 4
    yield 5
p=gen()
print(next(p))
print(next(p))
print(next(p))
print(next(p))
print(next(p))
print(next(p))
```
