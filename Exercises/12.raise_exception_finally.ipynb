```python
# Custom Exception
class IsMajor(Exception):
    def __init__(self, message):
        self.message = message


# Person class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age


    def get_minor_age(self):
        try:
            if self.age < 18:
                raise IsMajor("Person is not major")
            else:
                print("Person is major")

        except IsMajor as e:
            print(e.message)

        finally:
            print("Finally block executed")


    def display_person(self):
        print("Name:", self.name)
        print("Age:", self.age)


# Object
p = Person("Vardhan", 21)

p.display_person()
p.get_minor_age()
```
