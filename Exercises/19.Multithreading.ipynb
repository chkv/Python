print total active threads in multithreaded code using threading.active_count()

```python
import time
import threading
from threading import Thread

def sleepMe(i):
    print("Thread %i will sleep." % i)
    time.sleep(5)
    print("Thread %i is awake" % i)

threads = []

for i in range(10):
    th = Thread(target=sleepMe, args=(i,))
    th.start()
    threads.append(th)

for th in threads:
    th.join()

```
