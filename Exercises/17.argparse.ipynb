Exercise: Commandline Argument Processing using argparse

Take subject marks as command line arguments
example:
-                   python3 cmd.py --physics 60 --chemistry 70 --maths 90

```python
import argparse


parser=argparse.ArgumentParser()
parser.add_argument("--physics",help="Physics Marks",choices=range(0,101),type=int,default=0)
parser.add_argument("--chemistry",help="Chemistry Marks",type=int,choices=range(0,101),default=0)
parser.add_argument("--maths",help="Maths Marks",type=int,choices=range(0,101),default=0)
args=parser.parse_args()

print(args.physics)
print(args.chemistry)
print(args.maths)

print("Result:",(args.physics+args.chemistry+args.maths)/3)
```
